```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models, callbacks

# Set random seeds
tf.random.set_seed(42)
np.random.seed(42)

def generate_data(n_samples=1000):
    """Generate synthetic non-linear data"""
    X1 = np.random.uniform(-5, 5, n_samples)
    X2 = np.random.uniform(-5, 5, n_samples)
    X3 = np.random.uniform(-5, 5, n_samples)

    # Non-linear function with 3 variables
    y = (2 * np.sin(X1) + 0.5 * X2**2 + 0.3 * X3**3 +
         np.cos(X1 * X2) + np.random.normal(0, 0.2, n_samples))

    X = np.column_stack((X1, X2, X3))
    return X.astype(np.float32), y.reshape(-1, 1).astype(np.float32)

def create_model():
    """Create model using Keras Sequential API"""
    model = models.Sequential([
        # Input Layer
        layers.Input(shape=(3,)),

        # First Hidden Layer
        layers.Dense(64, kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Dropout(0.2),

        # Second Hidden Layer
        layers.Dense(32, kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Dropout(0.2),

        # Third Hidden Layer
        layers.Dense(16, kernel_initializer='he_normal'),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.Dropout(0.2),

        # Output Layer
        layers.Dense(1)
    ])

    return model

def plot_training_history(history):
    """Plot training history"""
    plt.figure(figsize=(20, 5))

    # Plot training history
    plt.subplot(131)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.yscale('log')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_predictions(model, X_test, y_test):
    """Plot prediction results"""
    plt.figure(figsize=(20, 5))

    # Get predictions
    y_pred = model.predict(X_test)

    # Plot 4D scatter
    ax = plt.subplot(132, projection='3d')
    scatter = ax.scatter(X_test[:, 0],
                        X_test[:, 1],
                        X_test[:, 2],
                        c=y_pred.ravel(),
                        cmap='viridis')
    plt.colorbar(scatter)
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('X3')
    plt.title('4D Scatter Plot\n(Color represents predicted value)')

    # Plot actual vs predicted
    plt.subplot(133)
    plt.scatter(y_test, y_pred, alpha=0.5)
    plt.plot([y_test.min(), y_test.max()],
             [y_test.min(), y_test.max()],
             'r--')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title('Actual vs Predicted Values')

    plt.tight_layout()
    plt.show()

def main():
    # Generate datasets
    X_train, y_train = generate_data(1000)
    X_val, y_val = generate_data(200)
    X_test, y_test = generate_data(200)

    # Create model
    model = create_model()

    # Compile model
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )

    # Model summary
    model.summary()

    # Create callbacks
    callbacks_list = [
        callbacks.EarlyStopping(
            monitor='val_loss',
            patience=20,
            restore_best_weights=True
        ),
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.1,
            patience=10,
            min_lr=1e-6
        ),
        callbacks.ModelCheckpoint(
            'best_model.h5',
            monitor='val_loss',
            save_best_only=True
        )
    ]

    # Train model
    print("\nTraining model...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=100,
        batch_size=32,
        callbacks=callbacks_list,
        verbose=1
    )

    # Plot training history
    plot_training_history(history)

    # Plot predictions
    plot_predictions(model, X_test, y_test)

    # Test predictions
    test_points = np.array([
        [-4, -4, -4],
        [-2, -2, -2],
        [0, 0, 0],
        [2, 2, 2],
        [4, 4, 4]
    ], dtype=np.float32)

    predictions = model.predict(test_points)

    print("\nTest Predictions:")
    for i, (x1, x2, x3) in enumerate(test_points):
        actual = 2 * np.sin(x1) + 0.5 * x2**2 + 0.3 * x3**3 + np.cos(x1 * x2)
        pred = predictions[i][0]
        print(f"Input: ({x1:2.0f}, {x2:2.0f}, {x3:2.0f})")
        print(f"Prediction: {pred:.3f}")
        print(f"Actual: {actual:.3f}")
        print(f"Difference: {abs(pred - actual):.3f}\n")

    # Model evaluation
    test_loss = model.evaluate(X_test, y_test, verbose=0)
    print(f"\nTest Loss: {test_loss[0]:.4f}")
    print(f"Test MAE: {test_loss[1]:.4f}")

if __name__ == "__main__":
    main()
```
